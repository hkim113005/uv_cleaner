# Import libraries and set constants
Import tensorflow and add relavent directories to the path variable.

```python
import os
import sys

import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
import cv2

# Adding needed directories to path 
base = os.path.join(os.path.expanduser("~"), "Downloads", "uv_cleaner", "human_detection")
sys.path.append(os.path.join(base))
sys.path.append(os.path.join(base, "image_detection"))

# Importing helpers
from helpers.image_helpers import *
from helpers.train_helpers import *

# Setting directories to save the model
tf_model_dir = os.path.join(base, "models")
tf_model_file = os.path.join(tf_model_dir, "saved_model.pb")
tf_properties_file = os.path.join(tf_model_dir, "saved_model.json")

# Set dataset location
dataset_base = os.path.join(base, "datasets")
dataset_dir = os.path.join(dataset_base, "data")

# Set images location
img_dir = os.path.join(dataset_base, "images")

# Size of images
input_size = (32, 24)
```

# Create datasets from images
Using images collected create datasets for ease of use.

```python
images, labels, classes = load_images(img_dir, resize=input_size, color_convert=cv2.COLOR_BGR2RGB)

labels = tf.keras.utils.to_categorical(labels)

train_images, test_images, train_labels, test_labels = train_test_split(
    np.array(images), np.array(labels), test_size=0.3
)
```

# Plot train images
Plotting sample of train images.

```python
plot_images_grid(
    images=train_images,
    labels=[np.argmax(label) for label in train_labels],
    classes=classes,
    rows=5,
    cols=5
)
```

# Declare and compile the model
## Hyperparameters
### Layers:
* 2D Convolution layer: 32 filters, (3, 3) kernel size, activation ReLU.
* 2D Max Pooling: (2, 2) pool size.
* Flatten.
* Dense layer: 256 nodes, activatino ReLU.
* Dropout: 0.2.
* Dense layer: 256 nodes, activation ReLU.
* Dropout: 0.1.
* Output layer: number of classes.
### Compilation:
* Loss function: categorical crossentropy.
* Optimzer: adam.
* Metrics: accuracy.

```python
model = keras.Sequential([
    keras.layers.Conv2D(
        32, 
        (3, 3), 
        activation="relu", 
        input_shape=train_images[0].shape
    ),
    keras.layers.MaxPooling2D(pool_size=(2, 2)),
    
    keras.layers.Flatten(),
    
    keras.layers.Dense(256, activation="relu"),
    keras.layers.Dropout(0.2),
    
    keras.layers.Dense(256, activation="relu"),
    keras.layers.Dropout(0.1),
    
    keras.layers.Dense(len(classes), activation="softmax")
])

model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)
```

# Train the model
Using EarlyStopping and validation for 20 epochs.

```python
model.fit(
    train_images,
    train_labels,
    epochs=20
)
```

# Evaluating the model

```python
results = model.evaluate(test_images, test_labels, verbose=2)

predictions = model.predict(test_images)
plot_results(
    images=test_images, 
    labels=[np.argmax(label) for label in test_labels], 
    classes=classes,
    predictions=predictions,
    rows=10, cols=4
)
```
